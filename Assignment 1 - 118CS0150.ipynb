```python
import numpy as np
import matplotlib.pyplot as plt
import math
from scipy.special import factorial
```

```python
def univariate_normal(a,n=1000):
    mu=a.mean()
    sigma=a.std()
    x=np.linspace(a.min(),a.max(),n)
    p=-1*pow((x-mu),2)/(2*pow(sigma,2))
    e=np.exp(p)
    y=1/(sigma*math.sqrt(2*math.pi))*e
    return x,y

a=np.arange(100)
x,y=univariate_normal(a,10000)
plt.title('Univariate Normal Distribution')
plt.plot(x,y)
```

```python
def exponential(x,lamda):
    y=[]
    for i in range(len(x)):
        if x[i]>0:
            y.append(lamda*math.exp(-1*lamda*x[i]))
        else:
            y.append(0)
    return y

x=np.random.rand(20)
x.sort()
lamda=2
y=exponential(x,lamda)
plt.title('Exponential Distribution')
plt.plot(x,y)
```

```python
def uniform(x,a,b):
    y=[]
    for i in range(len(x)):
        if x[i] in range(a,b+1):
            y.append(1/(b-a))
        else:
            y.append(0)
    return y

x=np.random.rand(20)
x.sort()
a=2
b=5
y=uniform(x,a,b)
plt.title('Uniform Distribution')
plt.plot(x,y)
```

```python
def gamma_dist(a,alpha,beta,n=1000):
    x=np.linspace(a.min(),a.max(),n)
    e=np.exp(-1/beta*x)
    coeff=(pow(beta,alpha)*math.gamma(alpha))
    p=np.power(a,alpha-1)
    y=coeff*p*e
    return x,y

a=np.arange(100)
alpha=0.5
beta=1
x,y=gamma_dist(a,alpha,beta,100)
plt.title('Gamma Distribution')
plt.plot(x,y)
```

```python
def binomial(x,p,n):
    y=[]
    q=1-p
    for i in range(n+1):
        if x[i]>0:
            coeff=math.factorial(n)/(math.factorial(x[i])*math.factorial(n-x[i]))
            y.append(coeff*pow(p,x[i])*pow(q,n-x[i]))
        else:
            y.append(0)
    return y

n=20
x=list(range(n+1))
p=0.4
y=binomial(x,p,n)
plt.title('Binomial Distribution')
plt.plot(x,y)
```

```python
def poisson(x,n):
    y=[]
    for i in range(n):
        if x[i]>0:
            y.append(pow(lamda,x[i])*math.exp(-1*lamda)/math.factorial(x[i]))
        else:
            y.append(0)
    return y

n=20
x=np.arange(n)
lamda=x.mean()
y=poisson(x,n)
plt.title('Poisson Distribution')
plt.plot(x,y)
```

```python
def beta_func(beta,alpha):
    ans=(math.gamma(alpha)*math.gamma(beta))/math.gamma(alpha+beta)
    return ans

def beta_dist(x,alpha,beta):
    y=[]
    func=beta_func(beta,alpha)
    for i in range(len(x)):
        y.append(pow(x[i],alpha-1)*pow(1-x[i],beta-1)/func)
    return y

x=list(range(2,21))
alpha=0.5
beta=0.5
y=beta_dist(x,alpha,beta)
plt.title('Beta Distribution')
plt.plot(x,y)
```
